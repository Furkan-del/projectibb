```python
# saatlik frekanstaki verilerin günlük frekansa çekilmesi,
# "DATE_TIME" kolonunun indeks olarak verilmesi ve data_daily adlı değişkene atanması:
```

```python
ibb_data=ibb_data.groupby('DATE_TIME',as_index=True).agg(sum)
ibb_data
```

```python
ibb_data.tail()
```

```python
ibb_data['DATE_TIME']=ibb_data.index
ibb_data
ibb_data.reset_index(drop=True)
```

```python
ibb_data['DATE_TIME']=pd.to_datetime(ibb_data['DATE_TIME'])
ibb_data.dtypes
```

```python
ibb_data.reset_index(drop=True)
data_daily = ibb_data.set_index("DATE_TIME").asfreq('D')
data_daily
```

```python
values=data_daily['NUMBER_OF_PASSENGER'].mean()
data_daily.fillna(value=values,inplace=True)
data_daily.isna().sum()
```

```python
data_daily['NUMBER_OF_PASSENGER']=data_daily['NUMBER_OF_PASSENGER'].astype(int)
```

```python
data_daily.plot(linestyle='solid',
        figsize=(15, 5),
        title='Zaman-Yolcu')

plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Zaman-Yolcu Graph')
plt.xlabel('Zaman')
plt.ylabel('Yolcu')
plt.plot(data_daily.NUMBER_OF_PASSENGER,color='red')
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
plt.gca().set(title='İstanbul Günlük Karayolu Ulaşımı Kullanan Kişi Sayıları')
sns.lineplot(data = data_daily, x=data_daily.index.values, y="NUMBER_OF_PASSENGER",label='Number Of PASSENGER',color='blue')
plt.xticks(rotation=45)
plt.show()
```

```python
## Bir takım deneyler
from prophet import Prophet
```

```python
data_daily
```

```python
df=data_daily.copy()
```

```python
df['DATE_TIME']=df.index
```

```python
df.index.name=None
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(df.columns[0],inplace=True,axis=1)
```

```python
df.columns=['y','ds']
```

```python
df
```

```python
model=Prophet()
model.fit(df)
```

```python
future=model.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast=model.predict(future)
forecast
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig1=model.plot(forecast)
plt.show()
```

```python
fig2 = model.plot_components(forecast)
```
