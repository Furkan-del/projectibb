```python
import os
import glob 
import pandas as pd
import numpy as np
path =r'C:\Users\90545\Desktop\repos\IBB-Project\datas' 
file_list = glob.glob(path + "/*.csv")
excl_list=[]

for file in file_list:
    excl_list.append(pd.read_csv(file))

excel_merged=pd.concat(excl_list,ignore_index=True)

excel_merged.head(10)
```
